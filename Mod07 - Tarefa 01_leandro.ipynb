#### 4) Qual variável é mais poderosa?

Considere as variáveis ```possui_email``` e ```posse_de_veiculo```. Faça uma tabela cruzada entre elas e responda qual delas te parece mais poderosa para prever a probabilidade de ```mau = 1```?

````python
email_vei = pd.crosstab(df_encoded['possui_email'], df_encoded['posse_de_veiculo'])
email_vei
````

````python
email_vei_pct = email_vei.div(email_vei.sum(axis = 1), axis = 0)
email_vei_pct
````

````python
email_mau = pd.crosstab(df_encoded['possui_email'], df_encoded['mau'])
email_mau_pct = email_mau.div(email_mau.sum(axis = 1), axis = 0)
email_mau_pct
````

````python
vei_mau = pd.crosstab(df_encoded['posse_de_veiculo'], df_encoded['mau'])
vei_mau_pct = vei_mau.div(vei_mau.sum(axis = 1), axis = 0)
vei_mau_pct
````

- 1) Com base nos dados, a porcentagem de maus pagadores para ambas variaveis indica independente da posse de veiculo ou email, mais de 97% dos clientes são bons pagadores.
- 2) Os dados também sugerem que talvez quem possui veículo tenha maior probabilidade de ser mau pagador, mas a diferença é muito baixa se comparada a quem possui email.
