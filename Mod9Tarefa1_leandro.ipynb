### 3. Qual o município que nasceu menos bebês em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
sinasc_raw.munResNome.value_counts().sort_values() 

# Castanheira aparece 32 vezes na df, logo, nasceram 32 crianças.
# No entanto, tem o objeto 'Município ignorado - RO', que optei por seguir o exemplo da df e ignorar
```

### 4. Qual o municipio que nasceu mais bebês no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



```python
sinasc_3 = sinasc_raw
```

```python
sinasc_3.loc[:,'DTNASC'] = pd.to_datetime(sinasc_raw['DTNASC'])
```

```python
sinasc_3[['ano','mês','dia']] = sinasc_3.DTNASC.dt.date.astype(str).str.split('-',expand=True)
sinasc_3.head()
```

```python
sinasc_3['mês'].unique()
```

```python
sinasc_3 = sinasc_3[sinasc_3['mês'] == '03'] # selecionando somente o mês de março que é '03'
```

```python
sinasc_3.munResNome.value_counts() # em março, Porto Velho teve a maior quantidade de nascimentos
```

Qual a quantidade de filhos vivos media, maxima, minima nesse municipio?

```python
sinasc_3 = sinasc_3[sinasc_3['munResNome'] == 'Porto Velho']
```

```python
sinasc_3.groupby('munResNome')['QTDFILVIVO'].agg([('mínima','min'),('média','mean'),('máxima', 'max')])
```

Qual a idade media, maxima, minima dos pais nesse municipio?

```python
sinasc_3.groupby('munResNome')['IDADEPAI'].agg([('mínima','min'),('média','mean'),('máxima', 'max')])
```

####INSIGHTS

1- Observa-se que a capital de RO, Porto Velho, registra a maior quantidade de nascimentos no ano de 2019, um dado que era o esperado para a maioria dos estados brasileiros

2- A média de idade dos pais em RO é de 31 anos e das mães 26. Em ambas cidades, Porto Velho com a maior quantidade de nascidos e Castanheiras com a menor quantidade de nascidos, os valores encontram-se próximos a essa tendencia central.

3- No entanto, a idade mínima das mães em Porto Velho é de 12 anos. Do ponto de vista jurídico, está informação revela um crime perpetrado pela cidade, na medida em que crianças abaixo de 16 anos não podem ser emancipadas, ou mesmo gozar de direitos e deveres da maioridade. De acordo com o ECA, estas meninas são consideradas ainda crianças na fase de pré adolescencia. O que revela um crime de abuso de incapaz na cidade. O comum é encontrar casos como esses nas cidade interioranas como é o caso de Castanheira pela má distribuição da informação referentes a educação sexual na cidade.

4- A quantidade de abandono paterno no ano de 2019 em RO foi em torno de 71%. Considerávelmente superior à porcentagem nacional para mesmo ano que é de 6,15%.

5- A maior parte das crianças, independente do nível educacional da mãe, nascem com raça Parda.

6- Em cidades mais distantes das metrópoles, como é o caso das cidades do estado de Rondônia, o parto cesareano tende a ter um número menor. Mas ao que tudo indica, em todo o Estado há uma tendencia maior em partos de cesareana.
