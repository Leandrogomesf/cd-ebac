####  2) Melhorando a interação com o API
Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?
4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

````python
#1) Seu código aqui

def pvl_brasil(UF = str, tipo_interessado = str):
    
    import pandas as pd
    import requests
    import datetime as dt
    
    # Linha para controle da UF
    
    try:
        UF = UF.upper()
    
    except:
        print('ERRO: Digite a sigla da UF em letras maiúsculas')
    
    # Linha para controle do tipo_interessado
    
   
    if (tipo_interessado == 'município' or 
        tipo_interessado =='municipio' or 
        tipo_interessado =='Município' or
        tipo_interessado == 'Municipio'):
        
        tipo_interessado = 'Munic%C3%ADpio'    
        
    elif (tipo_interessado == 'Estado' or tipo_interessado == 'estado'): 
        tipo_interessado = 'Estado'
    
    else:
        print('ERRO: Selecione Município ou Estado')
    
    try:
        url = f'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={UF}&tipo_interessado={tipo_interessado}'
        j_data = requests.get(url).json()
        df = pd.DataFrame(j_data['items'])
        df['data_status'] = pd.to_datetime(df['data_status'], dayfirst = True)
        
    except:
        print('ERRO: Não foi possível obter os dados. \n')
        print('Revise os argumentos de entrada.')
    
    return df
````

````python
# 2) Seu código aqui - Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com status em 'Arquivado por decurso de prazo' estão registradas?

pvl_MG = pvl_brasil('mg', 'estado')
````

````python
pvl_MG[pvl_MG['status'] == 'Arquivado por decurso de prazo'].nunique()
````

````python
# 3) Seu código aqui - Qual é o município da Bahia com mais solicitações deferidas?

pvl_BA = pvl_brasil('ba', 'municipio')
````

````python
pvl_BA.head()
````

````python
pvl_BA[pvl_BA['status'] == 'Deferido'].interessado.value_counts().head(1)
````

````python
# 4) Seu código aqui - Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

pvl_BA_estado = pvl_brasil('ba', 'estado')

pvl_BA_estado.to_csv('pvl_BA_estado.csv', index= False)
````
